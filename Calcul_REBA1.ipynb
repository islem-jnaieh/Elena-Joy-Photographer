```python
import csv

def calculate_reba_score(upper_arm, lower_arm, wrist, neck, trunk, leg, load, task_type):
    # Les coefficients de pondération pour chaque paramètre
    upper_arm_weights = [1, 2, 2, 2, 3]
    lower_arm_weights = [1, 2, 2, 2, 3]
    wrist_weights = [1, 2, 2, 2, 3]
    neck_weights = [1, 2, 2, 2, 3]
    trunk_weights = [1, 2, 2, 2, 3]
    leg_weights = [1, 2, 2, 2, 3]
    load_weights = [1, 2, 3, 3]
    task_type_weights = [1, 2, 3, 3]

    # Calcul du score pour chaque paramètre
    upper_arm_score = sum([a * b for a, b in zip(upper_arm, upper_arm_weights)])
    lower_arm_score = sum([a * b for a, b in zip(lower_arm, lower_arm_weights)])
    wrist_score = sum([a * b for a, b in zip(wrist, wrist_weights)])
    neck_score = sum([a * b for a, b in zip(neck, neck_weights)])
    trunk_score = sum([a * b for a, b in zip(trunk, trunk_weights)])
    leg_score = sum([a * b for a, b in zip(leg, leg_weights)])
    load_score = sum([a * b for a, b in zip(load, load_weights)])
    task_type_score = sum([a * b for a, b in zip(task_type, task_type_weights)])

    # Calcul du score total REBA
    reba_score = upper_arm_score + lower_arm_score + wrist_score + neck_score + trunk_score + leg_score + load_score + task_type_score

    return reba_score

def read_csv_file(filename):
    data = []
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip header
        for row in csv_reader:
            data.append(list(map(float, row)))
    return data

def main():
    filename = "donnees_posture.csv"  # Nom du fichier CSV
    posture_data = read_csv_file(filename)

    total_reba_score = 0
    for posture in posture_data:
        upper_arm = posture[0:5]
        lower_arm = posture[5:10]
        wrist = posture[10:15]
        neck = posture[15:20]
        trunk = posture[20:25]
        leg = posture[25:30]
        load = posture[30:34]
        task_type = posture[34:38]

        reba_score = calculate_reba_score(upper_arm, lower_arm, wrist, neck, trunk, leg, load, task_type)
        total_reba_score += reba_score

    average_reba_score = total_reba_score / len(posture_data)
    print("Le score REBA moyen est :", average_reba_score)

if __name__ == "__main__":
    main()
```
